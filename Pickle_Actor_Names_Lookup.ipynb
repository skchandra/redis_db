```python
import redis
r2 = redis.StrictRedis(host='localhost', port=6379, db=1)
r3 = redis.StrictRedis(host='localhost', port=6379, db=2)
pipe3 = r3.pipeline()
pipe2 = r2.pipeline()
```

```python
import pickle

# with open('actors_ids2.pickle', 'wb') as handle:
#   pickle.dump(new_actors_ids, handle)

with open('actors_ids2.pickle', 'rb') as handle:
  new_actor_ids = pickle.load(handle)

print new_actor_ids
```

```python
without_space_ids = {}
for key,value in new_actor_ids.iteritems():
    try:
        new_value = str(value)
        new_new_value = new_value.replace(" ", "")
        without_space_ids[key] = new_new_value
    except:
        new_value = value
        new_new_value = new_value.replace(" ", "")
        without_space_ids[key] = new_new_value
without_space_ids
```

```python
for key, value in new_actor_ids.iteritems():
    pipe2.execute();
    pipe2.set(key, value)
```

```python
for key, value in without_space_ids.iteritems():
    pipe3.execute();
    pipe3.set(value, key)
```
